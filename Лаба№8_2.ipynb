# Задача 1: Парсер складских остатков (CSV -> Подсчет)
У нас есть файл inventory.csv, в котором хранятся данные о товарах на складе. Формат записей: Товар,Количество,Цена.
Первая строка в файле — это заголовки столбцов, их считать не нужно.

**Ваша задача:**

1. Открыть файл и прочитать все строки.

2. Пропустить первую строку (заголовок). Подсказка: можно использовать срез lines[1:].

3. Пройтись циклом по оставшимся строкам. Очистить их от переносов строки (`.strip()`) и разбить по запятой (`.split(",")`).

4. Посчитать общую стоимость всех товаров на складе (для каждой строки умножить Количество на Цену и прибавить к общей сумме). Не забудьте перевести строки в числа с помощью `int()`

5. Вывести итоговую сумму на экран.

```python
total_value = 0

with open("inventory.csv", "r", encoding="utf-8") as file:
    lines = file.readlines()

for line in lines[1:]:
    parts = line.strip().split(",")
    quantity = int(parts[1])
    price = int(parts[2])
    total_value += quantity * price

print(f"Общая стоимость товаров на складе: {total_value} руб.")
```

# Задача 2: Конвертер базы данных (TXT -> JSON)
Вам передали выгрузку из старой базы данных в файле `employees.txt`. Данные записаны неудобно — через разделитель | (прямая черта).
Пример строки: Иван Иванов|Бухгалтерия|80000.

Современные системы общаются через `JSON`. Вам нужно написать скрипт-конвертер.

Ваша задача:

1. Создать пустой список `staff_list`.

2. Открыть файл `employees.txt` и прочитать его построчно.

3. В цикле разбить каждую строку по символу |.

4. Из полученных трех элементов (имя, отдел, зарплата) собрать словарь вида:
`{"name": "Иван Иванов", "department": "Бухгалтерия", "salary": 80000}`.

5. Добавить этот словарь в список `staff_list`.

6. После завершения цикла открыть новый файл `staff.json` в режиме записи ("w") и сохранить туда весь список `staff_list` с помощью json.dump(). Используйте параметр `indent=4` для красоты кода.

```python
import json

staff_list = []

with open("employees.txt", "r", encoding="utf-8") as file:
    for line in file:
        name, department, salary = line.strip().split("|")
        employee_dict = {
            "name": name,
            "department": department,
            "salary": int(salary)
        }
        staff_list.append(employee_dict)

with open("staff.json", "w", encoding="utf-8") as json_file:
    json.dump(staff_list, json_file, indent=4, ensure_ascii=False)

print("Конвертация завершена. Проверьте файл staff.json!")
```

# Задача 3: Безопасный калькулятор (try-except в цикле)
Данные из интернета или от пользователей редко бывают идеальными. В списке ниже хранятся числовые показатели с датчиков, переданные в виде строк. Но из-за сбоя связи некоторые значения испорчены (пришел текст ошибка и пустота None).

Дан список: `raw_data = ["10", "25", "ошибка", "40", "None", "15"]`

Ваша задача:

1. Создать переменную `total_sum = 0`.

2. Пройтись циклом for по списку `raw_data`.

3. Внутри цикла попытаться перевести текущий элемент в число (`int()`) и прибавить к `total_sum`.

4. Обернуть это действие в блок `try-except`. Если возникает `ValueError` (невозможно превратить текст в число), выведите сообщение формата: "Пропущено некорректное значение: `[значение]`" и позвольте циклу идти дальше.

5. Вывести итоговую сумму.

```python
raw_data = ["10", "25", "ошибка", "40", "None", "15"]
total_sum = 0

for item in raw_data:
    try:
        total_sum += int(item)
    except ValueError:
        print(f"Пропущено некорректное значение: {item}")

print(f"Итоговая сумма валидных данных: {total_sum}")
```
